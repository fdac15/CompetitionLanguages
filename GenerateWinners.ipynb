```python
import requests, json, sys, time, re
import operator
from bs4 import BeautifulSoup
def cleanHtml (html):
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
    return soup .get_text()

    
number = 1
myLinks = []
words = []
total = 0;
word = ""
Div1 = {}
Div2 = {}
f = open('Div1Winners.txt', 'r+')
f2 = open('Div2Winners.txt', 'r+')

while(number):
    URL = "https://community.topcoder.com/tc?module=SrmDivisionWins&sc=&sd=&nr=200&sr=" + str(number)
    number += 200
    r = requests .get(URL)
    t = r.content
    t = cleanHtml(t).lower()
    #print(t)
    
    if (number>1500):
        break;
        
    words = []
    #Create an array of the words in the file
    for letter in t:
        if(letter != ' ' and letter != '\n' ):
            word += letter
        else:
            if(word != ""):
                words.append(word)
            word = ""
    print(number)
    for i in range(len(words)):
        if(words[i] == "srm"):
            total += 1;
            k = 0;
            if(words[i + 2] in Div1):
                Div1[words[i+2]] += 1;
            else:
                Div1[words[i+2]] = 1;
            if(words[i + 2] == "n/a"):
                k = 1;
            if(words[i + 4] == "srm" or words[i + 4] == "1"):
                continue;
            if(words[i + 4 - k] in Div2):
                Div2[words[i+4 - k]] += 1;
            else:
                Div2[words[i+4 - k]] = 1;
            #Div2[words[i+4]] += 1;
    #print(Div1)
    #print(Div2)

sorted_Div1 = sorted(Div1.items(), key=operator.itemgetter(1))
print(sorted_Div1)
for value in sorted_Div1:
    print(value[1]);
    f.write(str(value[1]))
    f.write("\n")


print()
sorted_Div2 = sorted(Div2.items(), key=operator.itemgetter(1))
print(sorted_Div2)
for value in sorted_Div2:
    print(str(value[1]));
    f2.write((str(value[1])))
    f2.write("\n")
print(total)
```

```python
import json
countries = {};
countries2 = {};
tally = {}
def get_user_data(profile_url):
    # let it redirect us to the page with the user's handle in it
    redirected = requests.get(profile_url).url
    try:
        # get the handle from the url
        handle = re.search(r'/members/([^;]+)', redirected).groups(1)[0]
    except:
        print("Failed")
        return None
    members_api = "https://api.topcoder.com/v3.0.0/members/"
    # use the handle to get json data from members api
    r = requests.get(members_api + handle)
    return r.json()
f2 = open('winnersCountriesDiv1', 'w+')
f3 = open('winnersCountriesDiv2', 'w+')
for link in myLinks:
    url = link;
    #print(url)

    result = (get_user_data(url))
    try:
        if(result == None):
            continue;
    #print(result)
        handle = "";
        handle = result['result']['content']['handle']
        print(result['result']['content']['handle'])
        if(handle in Div1 and (handle not in Div2 or handle in tally)):
            print("test")
            if(result['result']['content']['competitionCountryCode'] in countries):
                countries[result['result']['content']['competitionCountryCode']] += 1;
            else:
                countries[result['result']['content']['competitionCountryCode']] = 1;
        else:
            print("test2")
            if(result['result']['content']['competitionCountryCode'] in countries2):
                countries2[result['result']['content']['competitionCountryCode']] += 1;
            else:
                countries2[result['result']['content']['competitionCountryCode']] = 1;
            tally[handle] = 1;
    except:
        print("test3")
        continue
for key, value in countries.items() :
    print (key, value)
    f2.write(key + "   " + str(value) + "\n");
print()
for key, value in countries2.items():
    print (key, value)
    f3.write(key + "   " + str(value) + "\n");
```
